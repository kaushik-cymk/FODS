```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Sampath\Desktop\Kaushik files\inventory_dataset.csv")
df
```

# 1. Handling Null Values

```python
# Identify the columns with null values in the dataset
null_columns = df.isnull().sum()
null_columns = null_columns[null_columns > 0]
null_columns
```

```python
# Impute 'Purchase_Price' with median
median_purchase_price = df['Purchase_Price'].median()
df['Purchase_Price'].fillna(median_purchase_price, inplace=True)
median_purchase_price
```

```python
# Impute 'Sale_Price' with mean
mean_sale_price = df['Sale_Price'].mean()
df['Sale_Price'].fillna(mean_sale_price, inplace=True)
mean_sale_price
```

```python
# Drop remaining rows with null values
df.dropna(inplace=True)
df
```

# 2. Handling Outliers

```python
# Identify Outliers in the 'Sale_Price' Column Using the Interquartile Range (IQR) Method

import pandas as pd
import numpy as np

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['Sale_Price'].quantile(0.25)
Q3 = df['Sale_Price'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['Sale_Price'] < lower_bound) | (df['Sale_Price'] > upper_bound)]
outliers
```

```python
# Remove outliers from the dataset
df = df[(df['Sale_Price'] >= lower_bound) & (df['Sale_Price'] <= upper_bound)]
df
```

```python
# Normalize the 'Quantity_In_Stock' Column Using Min-Max Scaling After Removing Outliers
from sklearn.preprocessing import MinMaxScaler

# Initialize the Min-Max Scaler
scaler = MinMaxScaler()

# Normalize the 'Quantity_In_Stock' column
df['Quantity_In_Stock'] = scaler.fit_transform(df[['Quantity_In_Stock']])
df['Quantity_In_Stock']
```

# 3. Data Cleaning

```python
#  Convert the 'Quantity_In_Stock' Column to Integer Type After Imputation
df['Quantity_In_Stock'] = df['Quantity_In_Stock'].astype(int)
df['Quantity_In_Stock']
```

```python
# Encode 'Discontinued' column using binary encoding
df['Discontinued'] = df['Discontinued'].map({'Yes': 1, 'No': 0})
df['Discontinued']
```

```python
# Save the cleaned dataset to a new CSV file
df.to_csv('cleaned_dataset.csv', index=False)
df.to_csv
```
